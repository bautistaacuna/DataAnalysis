# Screen Time Data: Productivity and Attention Span

El dataset “Screen Time Data: Productivity and Attention Span” (https://www.kaggle.com/datasets/muhammadalirazazaidi/screen-time-data-productivity-and-attention-span?utm_source=chatgpt.com) reúne respuestas de encuesta de 200 participantes con el objetivo de analizar cómo el tiempo frente a pantallas se relaciona con la productividad y la capacidad de atención. Además de medir el uso diario de dispositivos digitales, el conjunto de datos incorpora variables demográficas y patrones de comportamiento que permiten estudiar posibles asociaciones entre hábitos digitales y desempeño cognitivo.

Columnas:
*   Age Group: grupo etario del participante.
*   Gender: género del participante.
*   Education Level: nivel educativo alcanzado.
*   Occupation: ocupación o actividad principal.
*   Average Screen Time: promedio diario de tiempo frente a pantallas.
*   Device: dispositivo preferido para el uso digital.
*   Screen Activity: actividad principal realizada en pantalla.
*   App Category: categoría de aplicaciones más utilizadas.
*   Screen Time Period: momento del día en que se concentra el uso de pantalla.
*   Environment: entorno de trabajo preferido.
*   Productivity: nivel de productividad autopercibido al realizar tareas basadas en pantalla.
*   Attention Span: duración de la atención autopercibida durante tareas digitales.
*   Work Strategy: estrategia de trabajo preferida (por ejemplo, multitarea o trabajo enfocado).
*   Notification Handling: forma en que el participante gestiona notificaciones.
*   Usage of Productivity Apps: indica si utiliza herramientas digitales para mejorar su productividad.

Este conjunto de datos permite explorar patrones de consumo digital y analizar cómo distintos perfiles demográficos, hábitos de uso y estrategias de trabajo podrían estar asociados con variaciones en la atención y la productividad. En las siguientes secciones se desarrollará un análisis exploratorio para identificar tendencias, relaciones y posibles implicancias del uso intensivo de dispositivos digitales en el rendimiento cognitivo.

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
# Se carga el dataset
df = pd.read_csv("data.csv")

# Se realiza una vista rápida para verificar los datos
df.head()
```

## 1. Preparación de los datos

```python
df.info()
```

```python
#Se realiza un conteo de valores nulos por variable en porcentajes
(df.isnull().mean() * 100).round(2)
```

```python
# Se observa que hay una columna innecesaria creada llamada unnamed 0, la cual se procede a eliminarse.
df = df.drop(columns=["Unnamed: 0"])

# Tambien se observan algunos valores nulos, pero al ser pocos los registros (200 personas), esto equivale a: Environment = 1% (2 personas), Work Strategy = 1.5% (3 personas), Notification Handling = 0.5% (1 persona)
# Se proceden a eliminar los registros que contienen elementos vacios
df = df.dropna()

# Confirmar
df.info()
```

```python
#Se realiza un conteo de valores nulos por variable en porcentajes
(df.isnull().mean() * 100).round(2)
```

```python
# Se ven los tipos de valores unicos que contiene cada columna
for col in df.columns:
    print(f"\nColumna: {col}")
    print(df[col].unique())
```

## 2.Transformación de Variables

```python
# Average Screen Time: Posee valores del tipo ['More than 10' '8-10' '4–6' '6–8' '2–4' 'Less than 2']. Se convierte a variable del tipo numerica basandose en el promedio de ambos valores de cada rango.
screen_map = {
    "Less than 2": 1,
    "2–4": 3,
    "4–6": 5,
    "6–8": 7,
    "8-10": 9,
    "More than 10": 11
}
df["Screen_Time_Hours"] = df["Average Screen Time"].map(screen_map)
```

```python
# Attention Span: Se convierte a una variable ordinal numerica ya que posee valores del tipo ['10–30 minutes' 'More than 1 hour' '30–60 minutes' 'Less than 10 minutes']

attention_map = {
    "Less than 10 minutes": 1,
    "10–30 minutes": 2,
    "30–60 minutes": 3,
    "More than 1 hour": 4
}
df["Attention_Score"] = df["Attention Span"].map(attention_map)
```

```python
# Productivity: Se convierte a variable ordinal.
productivity_map = {
    "Unproductive, i might not have completed the task and got carried away": 1,
    "Moderately productive": 2,
    "Extremely productive, i efficiently complete my tasks": 3
}

df["Productivity_Score"] = df["Productivity"].map(productivity_map)
```

```python
# Age Group: Se convierte a una categoria ordenada
age_order = ["Below 18", "18–24", "25–34", "35–44", "45 and above"]

df["Age Group"] = pd.Categorical(
    df["Age Group"],
    categories=age_order,
    ordered=True
)
```

```python
# Gemder: Se convierte a categoría simple
df["Gender"] = df["Gender"].astype("category")
```

```python
# Education Level: Se convierte a variable ordinal
edu_order = ["High school or below", "Undergraduate", "Graduate"]

df["Education Level"] = pd.Categorical(
    df["Education Level"],
    categories=edu_order,
    ordered=True
)
```

```python
# Ocupation: A categoría
df["Occupation"] = df["Occupation"].astype("category")
```

```python
# Device, Screen Activity, App Category, Screen Time Period: Todas a catergoricas nominales
cols_nominal = [
    "Device",
    "Screen Activity",
    "App Category",
    "Screen Time Period",
    "Environment",
    "Work Strategy",
    "Notification Handling"
]

for col in cols_nominal:
    df[col] = df[col].astype("category")
```

```python
# Usage of Productivity Apps: Se transfroma a Ordinales
apps_map = {
    "No, i do not use them": 0,
    "Yes, but i did not find them of any help": 1,
    "Yes, they are extremely helpful": 2
}

df["Productivity_App_Score"] = df["Usage of Productivity Apps"].map(apps_map)
```

```python
# Se revisan nuevamente los valores unicos de cada variable
for col in df.columns:
    print(f"\nColumna: {col}")
    print(df[col].unique())
```

Las variables claves con las que se van a trabajar son:

*   Screen_Time_Hours (promedio diario de tiempo frente a pantallas): 11 , 9 , 5 , 7 , 3 or 1 hours.
*   Attention_Score (duración de la atención autopercibida durante tareas digitales):
    *   "Less than 10 minutes": 1
    *   "10–30 minutes": 2
    *   "30–60 minutes": 3
    *   "More than 1 hour": 4
*   Productivity_Score (nivel de productividad autopercibido al realizar tareas basadas en pantalla):
    *   "Unproductive, i might not have completed the task and got carried away": 1
    *   "Moderately productive": 2,
    *   "Extremely productive, i efficiently complete my tasks": 3
*   Age Group (grupo etario del participante):
    *   Below 18 years old
    *   18–24 years old
    *   25–34 years old
    *   35–44 years old
    *   45 and above years old
*   Screen Time Period (momento del día en que se concentra el uso de pantalla):
    *   Afternoon (12 PM–6 PM)
    *   Evening (6 PM–10 PM)
    *   Late night (10 PM–6 AM)
    *   Morning (6 AM–12 PM)
*   Notification Handling (forma en que el participante gestiona notificaciones):
    *   Check them briefly and resume my work
    *   Ignore them until my task is completed
    *   Spend time interacting with the notifications
    *   Turn off notifications altogether

### 3.7. Barplot: Cantidad de personas por tiempo de uso de pantalla por dia.

```python
import plotly.express as px
import pandas as pd

# Orden explícito para categorías (para que el eje X salga “bien”)
age_order = ["Below 18", "18–24", "25–34", "35–44", "45 and above"]
screen_order = ["Less than 2", "2–4", "4–6", "6–8", "8-10", "More than 10"]
period_order = ["Morning (6 AM–12 PM)", "Afternoon (12 PM–6 PM)", "Evening (6 PM–10 PM)", "Late night (10 PM–6 AM)"]
```

```python
counts = (
    df["Average Screen Time"]
    .value_counts(normalize=True)
    .reindex(screen_order)
    .reset_index()
)

counts.columns = ["Average Screen Time", "Percentage"]
counts["Percentage"] = counts["Percentage"] * 100

fig1 = px.bar(
    counts,
    x="Average Screen Time",
    y="Percentage",
    text=counts["Percentage"].round(1).astype(str) + "%",
    title="Distribución de tiempo de pantalla diario (%)",
    color_discrete_sequence=["#1f77b4"]
)

fig1.update_traces(
    textposition="outside",
    textfont=dict(size=20)
)

fig1.update_layout(
    template="simple_white",
    title=dict(
        text="Distribución de tiempo de pantalla diario (%)",
        x=0.5,  # 🔹 centrar título
        font=dict(size=24)
    ),
    xaxis=dict(
        title="Horas de pantalla/día",
        title_font=dict(size=24),
        tickfont=dict(size=20)
    ),
    yaxis=dict(
        title="Porcentaje de personas",
        title_font=dict(size=24),
        tickfont=dict(size=20),
        range=[0, counts["Percentage"].max() + 5]
    )
)

fig1.show()

```

Conclusión:
*   La mayoría de los participantes reportan entre 4 y 8 horas diarias de uso de pantalla, lo que refleja una exposición digital elevada y normalizada en la vida cotidiana.

```python
notif_order = (
    df.groupby("Notification Handling")["Productivity_Score"]
    .mean()
    .sort_values()
    .index
    .tolist()
)

fig = px.box(
    df,
    x="Notification Handling",
    y="Productivity_Score",
    category_orders={"Notification Handling": notif_order},
    points="all",
    title="Productividad vs Manejo de Notificaciones",
    color_discrete_sequence=["#1f77b4"]
)


fig.update_layout(
    template="simple_white",
    title=dict(
        text="Productividad vs Manejo de Notificaciones",
        x=0.5,
        font=dict(size=24)
    ),
    xaxis=dict(
        title="Manejo de Notificaciones",
        title_font=dict(size=24),
        tickfont=dict(size=20)
    ),
    yaxis=dict(
        title="Productividad (1=bajo, 3=alto)",
        title_font=dict(size=24),
        tickfont=dict(size=20)
    )
)


fig.update_xaxes(
    ticktext=[
        "Interactúa con notificaciones",
        "Revisa brevemente y continúa",
        "Desactiva notificaciones",
        "Ignora hasta terminar"
    ],
    tickvals=[
        "Spend time interacting with the notifications",
        "Check them briefly and resume my work",
        "Turn off notifications altogether",
        "Ignore them until my task is completed"
    ],
    tickangle=0
)


fig.update_traces(marker=dict(size=6))

fig.show()
```

Conclusiones:
*   “Interactúa con notificaciones” muestra la peor distribución: Interactuar activamente con notificaciones se asocia con menor productividad autopercibida de forma consistente, no solo en promedio sino en distribución.
*   “Ignora hasta terminar” muestra la mejor consistencia: Proteger el foco hasta completar la tarea se asocia con mayor productividad de manera consistente en la muestra.
*   “Desactiva notificaciones” también muestra buen rendimiento: Esto refuerza que el control de interrupciones tiene efecto observable.
*   Revisa brevemente y continúa” queda en el medio: Las micro-interrupciones parecen tener un costo intermedio.
