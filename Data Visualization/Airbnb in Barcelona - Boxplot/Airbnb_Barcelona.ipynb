#### Primeras Conclusiones

1.   Precio (realSum)
*   Los alojamientos completos son los más caros, con un precio medio muy elevado (572 €), casi 3 veces más que una habitación privada y 5 veces más que una compartida.
*   Las diferencias entre media y mediana en alojamientos completos (572 vs 474) indican una **asimetría positiva**, típica de mercados donde existen propiedades de lujo con valores muy altos.
*   El valor máximo (6943 €) confirma la presencia de **outliers extremos**, que elevan artificialmente el promedio.
*   **La dispersión** (std = 556 €) **es muy alta en Entire home/apt**, mientras que es mucho menor en habitaciones privadas, lo que sugiere un mercado de alojamientos completos más heterogéneo

2.   Capacidad y número de habitaciones
*   Los **alojamientos completos suelen tener más dormitorios y mayor capacidad**, lo que también explica parte del incremento del precio.
*   Las habitaciones privadas y compartidas mantienen estructuras más homogéneas

3.   Limpieza y satisfacción del huésped
*   **Las habitaciones privadas obtienen los mejores puntajes tanto en limpieza como en satisfacción**
*   Esto puede deberse a: espacios más pequeños, más fáciles de mantener limpios; interacción más directa del huésped con el anfitrión, expectativas más claras (un cuarto es más fácil de evaluar que una casa completa).
*   Las habitaciones compartidas presentan los valores más bajos, aunque el tamaño muestral de 8 limita la interpretación.

4.   Ubicación: distancia al centro y al metro
*   **Los alojamientos completos tienden a ubicarse más cerca del centro.**
*  ** Las habitaciones compartidas están más alejadas**, lo que coincide con estrategias orientadas a viajeros que priorizan precio por sobre ubicación.
*   **Tanto alojamientos completos como privados están relativamente bien conectados al metro** (< 0.45 km).
*   Esto sugiere dos segmentos de mercado: Turismo premium/parejas/familias, viviendas completas más céntricas; Turismo económico y jóvenes,  habitaciones privadas/compartidas más alejadas.













## 4. Visualizaciones

3.   Gráficos
*   Crea un gráfico de boxplot para diferentes categorías.
*   Incluye una breve explicación de las estadísticas mostradas en los boxplot de cada categoria (q1, mediana, etc..) y algunas conclusiones sobre las diferencias de los distintos grupos.




```python
# 3.1. Boxplot Precio vs Tipo de Alojamiento
fig = px.box(
    bc_w,
    x="room_type",
    y="realSum",
    labels={
        "room_type": "Tipo de Alojamiento",
        "realSum": "Precio (€)"
    },
    title="Boxplot: Precio según Tipo de Alojamiento"
)

fig.show()
```

El boxplot revela que el precio se relaciona fuertemente con el tipo de alojamiento. Los alojamientos completos presentan un mercado mucho más heterogéneo y con presencia de propiedades de lujo que generan outliers extremos, mientras que las habitaciones privadas y compartidas mantienen un comportamiento más estable, simétrico y accesible.

```python
# 3.2. Boxplot Precio vs Superhost
fig = px.box(bc_w, x="host_is_superhost",
    y="realSum",
    labels={
        "host_is_superhost": "Es Superhost",
        "realSum": "Precio (€)"
    },
    title="Precio según si el anfitrión es Superhost"
)
fig.show()
fig.write_image("boxplot_precio_superhost.png")
```

*   No existe una diferencia marcada en el precio entre Superhosts y no-Superhosts:
A simple vista, las cajas (IQR), las medianas y los rangos de ambos grupos son bastante similares. Esto indica que ser Superhost no determina un precio significativamente mayor.
*   Ambos grupos presentan outliers, pero los no-superhosts tienen los más extremos. **Esto sugiere que los alojamientos premium, muy costosos o de lujo no necesariamente están gestionados por Superhosts.**
*   El precio típico (mediana) es similar, pero los extremos alteran el promedio en no-superhosts pero la presencia de outliers extremos en el grupo false podría elevar la media mucho más que la mediana.








```python
# 3.3. Boxplot Puntuación de Limpieza vs Tipo de alojamiento
fig = px.box(bc_w, x="room_type",
    y="cleanliness_rating",
    labels={
        "room_type": "Tipo de Alojamiento",
        "cleanliness_rating": "Puntuación de Limpieza (0 a 10)"
    },
    title="Limpieza según Tipo de Alojamiento"
)
fig.show()
fig.write_image("boxplot_limpieza_tipo.png")
```



*   Las habitaciones privadas tienen la puntuación de limpieza más alta y más
estable: La mediana es 10, el valor máximo posible, el rango intercuartílico (Q1–Q3) se mantiene entre 9 y 10, mostrando muy poca variabilidad; los outliers son pocos, moderados y se concentran entre 2 y 7, lo cual indica que la mayoría de estos alojamientos alcanzan un nivel de limpieza sobresaliente
*   Los alojamientos completos presentan mayor variabilidad en limpieza: La mediana es 9, similar a las privadas, pero el rango intercuartílico está entre 8 y 10. Esto puede deberse a:mayor tamaño del inmueble (baño, cocina, varias habitaciones), variabilidad en la gestión de limpieza según el anfitrión, alojamientos antiguos o menos mantenidos.
*   Las habitaciones compartidas tienen puntuaciones ligeramente inferiores: La mediana está entre 8 y 9 (visible en el rango de la caja), no se observan outliers extremos, pero esto también está influenciado por el muy bajo tamaño muestral (solo 8 registros).












```python
# 3.4. Distancia al centro de la ciudad vs Tipo de Alojamiento
fig = px.box(bc_w, x="room_type",
    y="dist",
    labels={
        "room_type": "Tipo de Alojamiento",
        "dist": "Distancia al centro de la ciudad (km)"
    },
    title="Distancia al Centro según Tipo de Alojamiento"
)
fig.show()
fig.write_image("boxplot_distancia_centro.png")
```



*   Los alojamientos completos (Entire home/apt) están más cerca del centro: La mediana está alrededor de 1.8–2 km, el IQR (Q1–Q3) es estrecho, mostrando valores bastante consistentes y sin valores extremos
*   Las habitaciones privadas muestran la mayor variabilidad en distancia: La mediana está alrededor de 1.7 km similar a los alojamientos completos, Sin embargo, la dispersión es mucho mayor. **Esto indica un segmento de habitaciones privadas ubicado tanto en zonas muy céntricas como muy alejadas.**
*   Las habitaciones compartidas son las más alejadas del centro: La mediana se ubica en torno a 4 km, desplazado hacia zonas más periféricas. No aparecen outliers hacia abajo lo que quiere decir que no hay habitaciones compartidas cerca del centro.





```python
# 3.5. Distancia al metro vs Tipo de Alojamiento
fig = px.box(bc_w, x="room_type",
    y="metro_dist",
    labels={
        "room_type": "Tipo de Alojamiento",
        "metro_dist": "Distancia a la estación de metro (km)"
    },
    title="Distancia al Metro según Tipo de Alojamiento"
)
fig.show()
fig.write_image("boxplot_distancia_metro.png")
```



*   **Los alojamientos completos y privados tienen accesibilidad similar** texto en negrita: La mediana está entre 0.3 y 0.4 km para ambas categorías, esto indica muy buena accesibilidad al transporte público en estas dos categorías.
*   Las habitaciones privadas tienen mayor dispersión:Hay numerosos outliers que llegan hasta los 2.4km, lo cual confirma nuevamente que el mercado de habitaciones privadas es muy variado en ubicación
*   **Las habitaciones compartidas están más alejadas del metro**: La mediana está claramente más alta, en torno a 0.6 km, El IQR también es mayor: entre 0.3 y 1 km. Los alojamientos compartidos tienen peor acceso al metro, de nuevo consistente con menor cercanía al centro y precios más bajos.

*   Los precios más altos se concentran entre 1 y 3 km del centro. Los precios más altos se concentran entre 1 y 3 km del centro.
*   Los alojamientos completos (Entire home/apt) son los más caros y los más céntricos
*   Las habitaciones privadas son más baratas y están distribuidas en toda la ciudad
*   Las habitaciones compartidas son las más económicas y más alejadas
*   La puntuación general del huésped (tamaño de la burbuja) no muestra una relación fuerte con el precio ni con la distancia








El tamaño de las burbujas depende de la variable "guest_satisfaction_overall", las cuales son todas muy similares a 100%, motivo por el cual no se aprecian variaciones.

Se elige "cleanliness_rating" en su lugar que posee mas variacion:

```python
#3.7. Scatter Plot Distancia al centro de la ciudad vs Precio, color (Tipo de Alojamiento), Tamaño de la burbuja (Puntuación Limpieza).
fig = px.scatter(bc_w,
    x="dist",
    y="realSum",
    log_y=True,
    color="room_type",
    size="cleanliness_rating",
    hover_name="room_type",
    labels={
        "dist": "Distancia al centro de la ciudad (km)",
        "realSum": "Precio (€)",
        "room_type": "Tipo de Alojamiento",
        "cleanliness_rating": "Puntuación de limpieza (0 a 10)"
    },
    title="Scatter Plot Distancia al centro de la ciudad vs Precio (Limpieza)"
)
fig.show()
fig.write_image("scatter_distancia_precio_limpieza.png")
```

A diferencia del scatter anterior, aqui el tamaño de la burbuja se refiere a la Puntuacion de la Limpieza del alojamiento. **La puntuación de limpieza no muestra una relación fuerte con el precio ni con la distancia**
*   Se mantiene bastante similar entre alojamientos caros y económicos, también entre zonas céntricas y periféricas. Incluso alojamientos lejanos pueden tener limpieza 10/10, y alojamientos caros pueden tener puntuaciones más bajas.
